```python
import numpy as np
import pandas as pd
import pylab as plt
import seaborn as sns
import scipy.stats as stats

%matplotlib inline

fs = 40
t = 40
```

```python
dfx = pd.read_csv('accel_whole_x.csv', header=None)
dfy = pd.read_csv('accel_whole_y.csv', header=None)

[dfx.shape, dfy.shape]
```

```python
dfx_norm = (dfx - dfx.mean()) / (dfx.max() - dfx.min())
dfy_norm = (dfy - dfy.mean()) / (dfy.max() - dfy.min())

[dfx_norm.shape, dfy_norm.shape]
```

```python
dfx_std = np.std(dfx)

fig = plt.figure(figsize=(20, 16))
plt.title('Eixo X', fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs)
plt.xlabel('Teste (Amostras)', fontsize=fs)
plt.ylabel('Desvio Padro da Posio da Plataforma ()', fontsize=fs)

axes = plt.gca()
axes.set_xlim([0,9])
axes.set_ylim([0.02,0.05])

plt.plot(dfx_std, lw=10)

fig.savefig('t_x.png', dpi=300, bbox_inches='tight')
```

```python
dfy_std = np.std(dfy)

fig = plt.figure(figsize=(20, 16))
plt.title('Eixo Y', fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs)
plt.xlabel('Teste (Amostras)', fontsize=fs)
plt.ylabel('Desvio Padro da Posio da Plataforma ()', fontsize=fs)

axes = plt.gca()
axes.set_xlim([0,9])
axes.set_ylim([0.02,0.05])

plt.plot(dfy_std, lw=10)

fig.savefig('t_y.png', dpi=300, bbox_inches='tight')
```
