```python
import numpy as np
import pandas as pd
import pylab as plt
import seaborn as sns

%matplotlib inline

fs = 40
t = 40
```

```python
dfx = pd.read_csv('accel_movingavg_x.csv', header=None)
dfy = pd.read_csv('accel_movingavg_y.csv', header=None)

[dfx.shape, dfy.shape]
```

```python
dfx_norm = (dfx - dfx.mean()) / (dfx.max() - dfx.min())
dfy_norm = (dfy - dfy.mean()) / (dfy.max() - dfy.min())

[dfx_norm.shape, dfy_norm.shape]
```

```python
dfx_var = np.var(dfx)

plt.figure(figsize=(20, 20))
plt.title('Eixo X', fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs)
plt.xlabel('Janela do Filtro de Média Móvel', fontsize=fs)
plt.ylabel('Variância(°)', fontsize=fs)

plt.plot((t, t), (dfx_var.min(), dfx_var.max()), 'r--', lw=5)
plt.plot(dfx_var, lw=10)
```

```python
dfy_var = np.var(dfy)

plt.figure(figsize=(20, 20))
plt.title('Eixo Y', fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs)
plt.xlabel('Janela do Filtro de Média Móvel', fontsize=fs)
plt.ylabel('Variância(°)', fontsize=fs)

plt.plot((t, t), (dfy_var.min(), dfy_var.max()), 'r--', lw=5)
plt.plot(dfy_var, lw=10)
```
