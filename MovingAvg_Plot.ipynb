```python
import numpy as np
import pandas as pd
import pylab as plt
import seaborn as sns

%matplotlib inline

fs = 40
t = 40
```

```python
dfx = pd.read_csv('accel_movingavg_x.csv', header=None)
dfy = pd.read_csv('accel_movingavg_y.csv', header=None)

[dfx.shape, dfy.shape]
```

```python
dfx_norm = (dfx - dfx.mean()) / (dfx.max() - dfx.min())
dfy_norm = (dfy - dfy.mean()) / (dfy.max() - dfy.min())

[dfx_norm.shape, dfy_norm.shape]
```

```python
dfx_std = np.std(dfx)

fig = plt.figure(figsize=(20, 16))
plt.title('Eixo X', fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs)
plt.xlabel('Janela do Filtro de Mdia Mvel (Amostras)', fontsize=fs)
plt.ylabel('Desvio Padro da Posio da Plataforma ()', fontsize=fs)

plt.plot((t, t), (dfx_std.min(), dfx_std.max()), 'r--', lw=5)
plt.plot(dfx_std, lw=10)

fig.savefig('ma_x.png', dpi=300, bbox_inches='tight')
```

```python
dfy_std = np.std(dfy)

fig = plt.figure(figsize=(20, 16))
plt.title('Eixo Y', fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs)
plt.xlabel('Janela do Filtro de Mdia Mvel (Amostras)', fontsize=fs)
plt.ylabel('Desvio Padro da Posio da Plataforma ()', fontsize=fs)

plt.plot((t, t), (dfy_std.min(), dfy_std.max()), 'r--', lw=5)
plt.plot(dfy_std, lw=10)

fig.savefig('ma_y.png', dpi=300, bbox_inches='tight')
```
