```python
import numpy as np
import pandas as pd
import pylab as plt
import seaborn as sns
import scipy.stats as stats


%matplotlib inline

fs = 40
t = 40
```

```python
dfx = pd.read_csv('accel_whole_x.csv', header=None)
dfy = pd.read_csv('accel_whole_y.csv', header=None)

dfx1d = dfx.values.reshape(1000000,1)
dfy1d = dfy.values.reshape(1000000,1)

[dfx.shape, dfy.shape]
```

```python
dfx_norm = (dfx - dfx.mean()) / (dfx.max() - dfx.min())
dfy_norm = (dfy - dfy.mean()) / (dfy.max() - dfy.min())

[dfx_norm.shape, dfy_norm.shape]
```

```python
dfx_var = np.var(dfx)

plt.figure(figsize=(20, 20))
plt.title('Eixo X', fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs)
plt.xlabel('Janela do Filtro de Média Móvel', fontsize=fs)
plt.ylabel('Variância(°)', fontsize=fs)

#plt.plot(dfx_var, lw=10)
hmean = np.mean(dfx1d)
hstd = np.std(dfx1d)
pdf = stats.norm.pdf(dfx1d, hmean, hstd)
#plt.plot(dfx1d, pdf) # including h here is crucial
```

```python
dfy_var = np.var(dfy)

plt.figure(figsize=(20, 20))
plt.title('Eixo Y', fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs)
plt.xlabel('Janela do Filtro de Média Móvel', fontsize=fs)
plt.ylabel('Variância(°)', fontsize=fs)

plt.plot(dfy_var, lw=10)
```
